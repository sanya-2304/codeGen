```python
from google.colab import userdata
sec_key = userdata.get('codeAssis')
```

```python
#without ui
print("Warm welcome from your personal code assistant")

from huggingface_hub import InferenceClient
client = InferenceClient(model="mistralai/Mistral-7B-Instruct-v0.3", token=sec_key)

while True:
    prompt = input("\nEnter your prompt (or type 'exit' to quit): ")
    if prompt.lower() == "exit":
        break

    # Generate code response
    response = client.text_generation(
        prompt,
        max_new_tokens=300,  # Limit tokens to avoid excessive output
        temperature=0.3,  # Lower temp for accuracy
        top_p=0.9  # Reduce randomness
    )

    print("\nGenerated Code:\n")
    print(response)
```

```python
from huggingface_hub import InferenceClient
import gradio as gr

# Initialize the client
client = InferenceClient(model="mistralai/Mistral-7B-Instruct-v0.3", token=sec_key)

# Function to generate code
def generate_code(prompt):
    response = client.text_generation(
        prompt,
        max_new_tokens=300,  # Limit tokens to avoid excessive output
        temperature=0.3,  # Lower temp for accuracy
        top_p=0.9  # Reduce randomness
    )
    return response

# Gradio UI
demo = gr.Interface(
    fn=generate_code,
    inputs=gr.Textbox(label="Enter your prompt"),
    outputs=gr.Textbox(label="Generated Code"),
    title="CodeGen",
    description="Enter a prompt to generate relevant code:",
    theme="compact"
)

# Launch the Gradio UI
demo.launch()
```
